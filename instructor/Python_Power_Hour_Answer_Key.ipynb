# Step 3: Storing Information тАФ Variables

A **variable** is like a labelled box where your program stores information.

```

```

Just remember two things:
- ЁЯУЭ Words and sentences go inside quotes: `"like this"`
- ЁЯФв Numbers go without quotes: `42` or `3.5`

You can print a variable just by putting its name inside `print()`.

### Naming your variables

Keep variable names simple. Two main rules:
- Use lowercase letters and underscores: `lucky_number`, `favourite_colour`

# Step 5: Making Choices тАФ `if` / `else`

Python runs every line top to bottom. But you can give it a fork in the road.

```python
if <something is true>:
    do this
else:
    do this instead
```

The most common check is `==`, which asks "is this equal to?"

> Watch out: `=` stores a value. `==` checks if two things match. Easy to mix up!

### Checking things

Here are the symbols you'll use to compare values:

| Symbol | What it means | Example |
|--------|--------------|---------|
| `==` | Is equal to | `mood == "happy"` |
| `!=` | Is NOT equal to | `colour != "red"` |
| `>` | Greater than | `score > 50` |

Each check gives back either `True` or `False`.

### ЁЯСйтАНЁЯТ╗ Code Along
Follow your instructor's lead on this one!

````python
# A simple if / else
weather = "sunny"

if weather == "sunny":
    print("Let's go outside! тШАя╕П")
else:
    print("Maybe stay in and code. ЁЯТ╗")
````

Want more than two choices? Add `elif` (short for "else if") between your `if` and `else`:

```python
if answer == "yes":
    print("Awesome!")
elif answer == "maybe":
    print("Think about it!")
else:
    print("That's okay too!")
```

````python
score = 75

if score > 90:
    print("Incredible! ЁЯПЖ")
elif score > 60:
    print("Nice work! ЁЯСН")
else:
    print("Keep going! ЁЯТк")
````

### ЁЯЪА Try it out! тАФ The Mood Machine ЁЯОн

Build a program that:
1. Asks the user how they're feeling
2. Responds differently depending on their answer

Use at least one `if`, one `elif`, and one `else`.

````python
# === The Mood Machine ===
# Ask the user how they feel
mood = input("How are you feeling today? (happy / tired / hungry): ")

# Respond based on their answer
if mood == "happy":
    print("Love it! Keep that energy! ЁЯМЯ")
elif mood == "tired":
    print("Hang in there! Almost snack break time ЁЯШ┤")
elif mood == "hungry":
    print("Same... let's finish this section first ЁЯНХ")
else:
    print("Interesting choice! I'll have to think about that one ЁЯдФ")
````

# The Secret Ingredient ЁЯО▓

Two things you need for the Magic 8-Ball:

A **list** stores a bunch of items together inside square brackets:

```python
answers = ["Yes!", "No way!", "Maybe...", "Definitely!"]
```

`import random` brings in Python's randomness toolbox. Then `random.choice()` picks one item from your list, different every time:

```python

random.choice(answers)
```

Run the cell below a few times and watch what happens.

````python
import random

# A list of possible answers
answers = ["Yes!", "No way!", "Maybe...", "Definitely!", "Ask me tomorrow..."]

# random.choice() picks one item at random тАФ run this a few times!
print(random.choice(answers))
print(random.choice(answers))
print(random.choice(answers))
````
